```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\jbwil\Downloads\IMBD.csv\IMBD.csv")

# Create a Bar graph for highest ratings by genre
plt.figure(figsize=(10, 6))
df_bar = df.groupby('genre')['rating'].mean().sort_values(ascending=False)[:10]
df_bar.plot(kind='bar', color='blue')
plt.title('Highest Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Highest Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# b. Identify the top 5 movies (highest ratings)
top_movies = df.nlargest(5, 'rating')[['movie', 'rating']]
print("Top 5 Movies:")
print(top_movies)

# c. Compare top 5 genres with a pie chart
top_genres = df['genre'].value_counts().head(5)
plt.figure(figsize=(8, 8))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Genres Distribution')
plt.axis('equal')
plt.show()
```
